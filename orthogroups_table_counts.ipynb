```python
import pandas as pd
import numpy as np
```

```python
# orthogroup data
df = pd.read_csv("Orthogroups.tsv", sep="\t")
```

```python
df.head()
```

```python
# drop duplicate genomes from orthogroups
df.drop(labels=["leptospirillum_ferriphilum","stenotrophomonas_maltophilia"], axis=1, inplace=True)
```

```python
# candidate genes data
candidates = pd.read_csv("candidategenes.csv", header=None)
```

```python
candidates.head()
```

```python
candidates.shape
```

```python
# drop duplicate candidate genes
candidates.drop_duplicates(inplace=True)
```

```python
candidates.shape
```

```python
# convert candidate genes df to a set
candidates = set(candidates.iloc[:,0].tolist())
```

```python
# filter df to only show orthogroups corresponding to the list of candidate genes
filterer = lambda x: ", ".join(y for y in x.split(", ") if y in candidates)
```

```python
# filtered dataframe output
df_out = (df.set_index("Orthogroup").iloc[:,8:]
          .fillna("")
          .applymap(filterer)
          .replace("", np.nan)
          .dropna(how="all"))
```

```python
df_out.head()
```

```python
df_out.shape
```

```python
# count overlaps
overlaps = [count.str.split(",").explode().count() for i, count in df_out.iterrows()]
```

```python
len(overlaps)
```

```python
# add back in the MAGs and Ac. genomes
df_out = pd.merge(df_out, df.set_index("Orthogroup").iloc[:,0:8], left_index=True, right_index=True, how="left")
```

```python
df_out.shape
```

```python
df_out.head()
```

```python
# count genes in organisms
gene_counts = pd.DataFrame(
    data = np.array([counts.str.split(",").str.len() for i, counts in df_out.iterrows()]).astype(int),
    columns = [str(name)+"_counts" for name in df_out.columns],
    index = df_out.index)
```

```python
gene_counts.shape
```

```python
gene_counts.head()
```

```python
df_final = pd.concat([df_out, gene_counts], axis=1).reset_index()
df_final.insert(1, "overlaps", overlaps)
```

```python
df_final.shape
```

```python
df_final.head()
```

```python
df_final.to_csv("orthogroups_table.csv", index=False)
```
